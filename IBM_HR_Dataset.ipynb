```python
def Box_plots(df):
    plt.title("Box Plot")
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 # Add a graph in each part
    sns.boxplot(df, ax=ax_box)
    sns.distplot(df, ax=ax_hist)
    ax_box.set(xlabel='')
```
